```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Create synthetic dataset
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Visualize the dataset
plt.scatter(X[:, 0], X[:, 1], s=30)
plt.title("Original Data")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
y_kmeans = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', label='Centroids', marker='X')
plt.title("K-Means Clustering")
plt.legend()
plt.show()
```

```python
dbscan = DBSCAN(eps=0.9, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='plasma')
plt.title("DBSCAN Clustering")
plt.show()
```

```python
agg = AgglomerativeClustering(n_clusters=4)
y_agg = agg.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_agg, cmap='Accent')
plt.title("Hierarchical Clustering (Agglomerative)")
plt.show()
```

```python
linked = linkage(X, method='ward')

plt.figure(figsize=(10, 6))
dendrogram(linked, truncate_mode='lastp', p=20)
plt.title("Dendrogram")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# Create moon-shaped data
X, y = make_moons(n_samples=300, noise=0.05, random_state=0)

# Visualize original moon data
plt.scatter(X[:, 0], X[:, 1], s=30)
plt.title("Moon-Shaped Dataset")
plt.show()
```

```python
  # DBSCAN clustering
dbscan = DBSCAN(eps=0.2, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

# Visualize DBSCAN clusters
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='Set1')
plt.title("DBSCAN on Moon Data")
plt.show()
```
